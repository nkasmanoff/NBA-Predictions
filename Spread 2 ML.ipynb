```python
def ML2Payout(ML,bet):
    """Convert Moneyline odds to a payout. 
    """
    
    if ML < 0: # - moneyline, 
        # PAYOUT = BET AMOUNT / (-1 *MONEYLINE ODDS / 100)

        payout = bet / (-1*ML/100)

    elif ML > 0:   #now for the underdog
        #PAYOUT = BET AMOUNT * ODDS / 100
        payout = (bet * ML) / 100

            
    else:
        payout = bet

    return payout
```
